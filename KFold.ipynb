### sklearn.model_selection.KFold
* class sklearn.model_selection.KFold(n_splits=5, *, shuffle=False, random_state=None)


```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features= iris.data # 피쳐 4개였음
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

# 5개의 폴드 세트로 분리하는 KFold 객체와 폴드 세트별
kfold = KFold(n_splits=5)
cv_accuracy = []
print('붓꽃 데이터 세트 크기:',features.shape[0] )
print(kfold)
```

```python
n_iter = 0
 # KFold객체의 split() 호출하면 폴드 별 학습용, 검증용 테스트의 로우 인덱스를 array로 반환
for train_index, test_index in kfold.split(features):
  # kfold.split()으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
  X_train, X_test = features[train_index], features[test_index]
  y_train, y_test = label[train_index], label[test_index]

  # 학습 및 예측
  dt_clf.fit(X_train, y_train)
  pred = dt_clf.predict(X_test)
  n_iter += 1

  # 반복 시 마다 정확도 측정
  accuracy = np.round(accuracy_score(y_test,pred),4)
  train_size=X_train.shape[0]
  test_size = X_test.shape[0]
  print('\n#{0} 교차검증 정확도 : {1}, 학습데이터 크기: {2}, 검증 데이터 크기: {3}'
        .format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스:{1}'.format(n_iter, test_index))
  cv_accuracy.append(accuracy)

print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
# 이렇게 해도됨
#f살려서 변수를 직접 넣어도 됨 . f'\n#{n_iter}교차검증 정확도 : {accuracy}, 학습데이터 크기: {train_size}, 검증 데이터 크기: {test_size}'

```

```python
## 참고 KFold의 단점 - 3으로나누면 안됨 - stratify하게 나눠줘야함
# regression 같은 거는 kfold걍 해도 되는데,  이런데이터들은 stratify 하게, 균형있게 나눠줘야

kfold = KFold(n_splits=3)
cv_accuracy = []
print('붓꽃 데이터 세트 크기:',features.shape[0] )
print(kfold)
```

```python
n_iter = 0
 # KFold객체의 split() 호출하면 폴드 별 학습용, 검증용 테스트의 로우 인덱스를 array로 반환
for train_index, test_index in kfold.split(features):
  # kfold.split()으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
  X_train, X_test = features[train_index], features[test_index]
  y_train, y_test = label[train_index], label[test_index]

  # 학습 및 예측
  dt_clf.fit(X_train, y_train)
  pred = dt_clf.predict(X_test)
  n_iter += 1

  # 반복 시 마다 정확도 측정
  accuracy = np.round(accuracy_score(y_test,pred),4)
  train_size=X_train.shape[0]
  test_size = X_test.shape[0]
  print('\n#{0} 교차검증 정확도 : {1}, 학습데이터 크기: {2}, 검증 데이터 크기: {3}'
        .format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스:{1}'.format(n_iter, test_index))
  cv_accuracy.append(accuracy)

print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))


# 0,1,2 이씅면  kfold를 3개로 하면 검증셋이 0부터 49 가 검증데이터니까 1,2로 훈련하고 레이블 0인걸로 검증하니까 1,2,밖에 예측못함 그니까 정확도 0
```
